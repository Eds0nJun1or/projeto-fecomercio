#### Consumindo API e enviando dados para o banco

```python
# Endpoint da API para transações Pix por município
url = "https://olinda.bcb.gov.br/olinda/servico/Pix_DadosAbertos/versao/v1/odata/TransacoesPixPorMunicipio(DataBase=@DataBase)?@DataBase='202101'&$top=1000&$format=json&$select=AnoMes,Municipio_Ibge,Municipio,Estado_Ibge,Estado,Sigla_Regiao,Regiao,VL_PagadorPF,QT_PagadorPF,VL_PagadorPJ,QT_PagadorPJ,VL_RecebedorPF,QT_RecebedorPF,VL_RecebedorPJ,QT_RecebedorPJ,QT_PES_PagadorPF,QT_PES_PagadorPJ,QT_PES_RecebedorPF,QT_PES_RecebedorPJ"

# Realizar a requisição
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    data = response.json()

    # Extrair e organizar os dados
    resultados = []
    for item in data['value']:
        ano_mes = item['AnoMes']
        municipio_ibge = item['Municipio_Ibge']
        municipio = item['Municipio']
        estado_ibge = item['Estado_Ibge']
        estado = item['Estado']
        sigla_regiao = item['Sigla_Regiao']
        regiao = item['Regiao']
        vl_pagador_pf = item['VL_PagadorPF']
        qt_pagador_pf = item['QT_PagadorPF']
        vl_pagador_pj = item['VL_PagadorPJ']
        qt_pagador_pj = item['QT_PagadorPJ']
        vl_recebedor_pf = item['VL_RecebedorPF']
        qt_recebedor_pf = item['QT_RecebedorPF']
        vl_recebedor_pj = item['VL_RecebedorPJ']
        qt_recebedor_pj = item['QT_RecebedorPJ']
        qt_pes_pagador_pf = item['QT_PES_PagadorPF']
        qt_pes_pagador_pj = item['QT_PES_PagadorPJ']
        qt_pes_recebedor_pf = item['QT_PES_RecebedorPF']
        qt_pes_recebedor_pj = item['QT_PES_RecebedorPJ']
        
        resultados.append({
            "AnoMes": ano_mes,
            "Municipio_IBGE": municipio_ibge,
            "Municipio": municipio,
            "Estado_IBGE": estado_ibge,
            "Estado": estado,
            "Sigla_Regiao": sigla_regiao,
            "Regiao": regiao,
            "VL_PagadorPF": vl_pagador_pf,
            "QT_PagadorPF": qt_pagador_pf,
            "VL_PagadorPJ": vl_pagador_pj,
            "QT_PagadorPJ": qt_pagador_pj,
            "VL_RecebedorPF": vl_recebedor_pf,
            "QT_RecebedorPF": qt_recebedor_pf,
            "VL_RecebedorPJ": vl_recebedor_pj,
            "QT_RecebedorPJ": qt_recebedor_pj,
            "QT_PES_PagadorPF": qt_pes_pagador_pf,
            "QT_PES_PagadorPJ": qt_pes_pagador_pj,
            "QT_PES_RecebedorPF": qt_pes_recebedor_pf,
            "QT_PES_RecebedorPJ": qt_pes_recebedor_pj
        })

    # Organizar os dados como necessário
    # Por exemplo, imprimir os resultados
    for resultado in resultados:
        print(resultado)
else:
    print("Erro:", response.status_code, response.text)
```

```python
# Converter para DataFrame
df = pd.DataFrame(resultados)
print(df)
```

```python
#tranformar dados em dicionário para insertir no banco de dados
data_dict = df.to_dict("records")
print(data_dict)
```

```python
# Inserindo os dados no MongoDB
collection.insert_many(data_dict)
```
